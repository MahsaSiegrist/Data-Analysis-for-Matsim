```python
import pandas as pd
```

# STEP 1 read the files

```python
df1 = pd.read_csv("replan_Immediate_information.txt", sep='	')
df2 = pd.read_csv("replan_replan_when_boardingDenied.txt", sep='	')
```

# STEP 2 just keep T2 in file one

```python
df1 = df1[df1.triggerType == 'T-2']
df2 = df2[df2.triggerType == 'T-2']
```

# STEP 3 Remove duplicates in each file

```python
df1 = df1.drop_duplicates(subset=['personId', 'triggerType',
       'currentStopName', 'transitLine', 'transitRoute', 'vehicle',
       'nextStopName', 'passengersSize', 'passengerCapacit'], keep='first') # hen we select unique users.
df2 = df2.drop_duplicates(subset=['personId', 'triggerType',
       'currentStopName', 'transitLine', 'transitRoute', 'vehicle',
       'nextStopName', 'passengersSize', 'passengerCapacit'], keep='first') # hen we select unique users.
```

```python
df1.personId.unique().shape
```

## Let's see the data

```python
df1['checker'] = 1
df2['checker'] = 1
```

## We select unique users who are in df1 and not in df2

```python
result1 = pd.merge(df1, df2, on='personId', how="left") # Make a left join, then clear the data.
result1 = result1[pd.isnull(result1.checker_y)] # Next, filter the data that does not have the right data.
result1u = result1.drop_duplicates(subset=['personId']) # hen we select unique users.
result1u.shape # printing resultat
```

## We select unique users who are in df1 and not in df2

```python
# We do the same thing as above
result2 = pd.merge(df2, df1, on='personId', how="left")
result2 = result2[pd.isnull(result2.checker_y)]
result2u = result2.drop_duplicates(subset=['personId'])
result2u.shape
```

## Merge two frames on the key field personId and get the intersection of both data sets

```python
# Merge all the values between the two data frames that are exactly the same (with inner)
df_all_same = pd.merge(df1, df2 , on='personId', how='inner').drop_duplicates('personId')
df_all_same
```

## Further, it turns out that each of these elements is contained in df1 and df2, we can check this:

```python
df_all_same.merge(df1, on='personId').drop_duplicates('personId')
```

```python
df_all_same.merge(df2, on='personId').drop_duplicates('personId')
```
